## Facetgrid, factorplot y lmplot

El concepto de los grÃ¡ficos mÃºltiples es ideal para datasets con muchas variables. Esto permite identificar tendencias rÃ¡pidamente comparando mÃºltiples grÃ¡ficos, uno al lado del otro, usando la misma escala y ejes. Estos plots son referidos como **trellis o lattice plot**. En Data Science, este concepto es normalmente llamado *__Faceting__*.
**FacetGrid** se encarga de organizar la data en forma que se disponga segÃºn filas, columnas y hue, y luego sea utilizada con el tipo apropiado de plot. ```FacetGrid()``` es muy flexible, pero es necesario usar un proceso de dos partes donde primero se definan las Facets y luego se mapee el tipo de plot. Por ejemplo, esta es una forma de usar ```FacetGrid()``` con boxplot:

````python
g = sns.FacetGrid(df, col="var categorica a plotear") #PRIMERO, definimos los facets
g.map(sns.boxplot, 'var continua a usar',       
      order=[*orden de los valores de la columna que utilizo*]) #SEGUNDO, mapeamos los facets
````

```factorplot()``` **es una manera sencilla de utilizar ```FacetGrid()``` sobre variables categÃ³ricas**. Combina los procesos de faceting y mapping en una sola linea.

````python
sns.factorplot(x='var x continua', data=df,               
               col="var col categorica", kind='box') #boxplot para este caso
````

## PairGrid y pairplot

Son similares a FacetGrid, factorplot y lmplot porque nos permiten ver tendencias a travÃ©s de diferentes columnas de data. La diferencia es que **solo definimos las columnas de data que queremos comparar**.
El procedimiento para hacer el ```PairGrid()``` es similar al que se usa en ```FacetGrid()```, primero definis las Facets y luego mapeas el tipo de plot de la grilla.
__Lo interesante de este codigo es que no tenemos que definir los parametros de fila y columna.__

````python
g = sns.PairGrid(df, vars=["var 1","var 2"])
g = g.map(plt.scatter)
````



El grÃ¡fico en linea recta, sin embargo, no nos dice mucho al respecto. Podemos cambiar el tipo de grÃ¡fico de la siguiente manera:

````python
g = sns.PairGrid(df, vars=["var 1","var 2"])
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
#map_diag y map_offdiag definen los grÃ¡ficos de las diagonales.
````

### AÃ±adir lineas de referencia con ```axvline()```

Este mÃ©todo permite aÃ±adir lineas de referencia en el eje x. Veamos un ejemplo donde se insertan dos referencias: una con la mediana y otra con la media

````python
# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=df['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
````
